### Definición de Análisis de Texto
El análisis de texto es una rama de la ciencia de datos que se centra en la extracción de información útil y significativa a partir de datos de texto no estructurados.

### Importancia del Análisis de Texto
Permite obtener conocimientos a partir de grandes volúmenes de texto, como redes sociales, documentos empresariales, correos electrónicos, etc.
Facilita la automatización de tareas como la clasificación de documentos, el análisis de sentimientos y la extracción de información.

### Aplicaciones del Análisis de Texto

#### Aplicaciones Principales
- Clasificación de documentos
- Análisis de sentimientos
- Extracción de información
- Generación de resúmenes automáticos
- Traducción automática
- Chatbots y asistentes virtuales

#### Ejemplos de Casos de Uso
- Análisis de opiniones en redes sociales para entender la percepción del cliente.
- Clasificación de correos electrónicos como spam o no spam.
- Extracción de entidades (nombres, lugares, fechas) de documentos legales.


### Componentes del Análisis de Texto
#### Preprocesamiento de Texto
- Limpieza de datos: eliminación de signos de puntuación, números, palabras vacías (_stopwords_).
- Tokenización: dividir el texto en unidades más pequeñas (tokens).

#### Representación de Texto
- Bolsa de palabras (Bag of Words): representación vectorial basada en la frecuencia de palabras.
- Modelos de incrustación de palabras (Word Embeddings): representación densa de palabras en un espacio vectorial.

### Preprocesamiento

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
texto = "¡Este es un ejemplo de preprocesamiento de texto en Python!"

# Eliminar signos de puntuación y convertir a minúsculas
texto_limpio = re.sub(r'[^\w\s]', '', texto.lower())

texto_limpio
```

```python
# Eliminar stopwords
stop_words = set(stopwords.words('spanish'))
list(stop_words)[:20]
```

```python
tokens = word_tokenize(texto_limpio)
tokens
```

```python
t = "no era el más honesto no el más piadoso pero era un hombre valiente."
tokens = word_tokenize(t)
texto_filtrado = [word for word in tokens if word not in stop_words]

texto_filtrado
```

```python
for word in ['ejemplo', 'po', 'cachai', 'weon', 'cuático', 'poas']:
    stop_words.add(word)

texto_filtrado = [word for word in tokens if word not in stop_words]

texto_filtrado
```

```python
def tokens(txt, idioma = 'english'):
    texto_limpio = re.sub(r'[^\w\s]', '', txt.lower())

    stop_words = set(stopwords.words(idioma))
    tokens = word_tokenize(texto_limpio)
    texto_filtrado = [word for word in tokens if word not in stop_words]
    return texto_filtrado
```

```python
map_libro = list(map(tokens, libro_filitrado))
```

### Eliminación de desinencias ([_stemming_](https://www.nltk.org/howto/stem.html))

[RAE](https://dle.rae.es/desinencia): Morfema flexivo pospuesto a la raíz, especialmente la de un verbo.

>¿Cuál es la raíz de las siguientes palabras? <br>
Correr, corríamos, corran, correrán, corrieran, corriendo.

> ¿Cuáles son las desinencias?

Características:
- Simplifica
- Método rápido
- **No toma en cuenta la gramática**

Hay diferentes algoritmos para hacer _stemming_, el más común por lo sencillo es [PorterStemmer](https://tartarus.org/martin/PorterStemmer/python.txt).

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

texto = "running runs runner runners"

palabras = word_tokenize(texto)

stemmed_words = [ps.stem(palabra) for palabra in palabras]
stemmed_words
```

Otros son más agresivos.

```python
from nltk.stem import LancasterStemmer

ls = LancasterStemmer()

stemmed_words = [ls.stem(palabra) for palabra in palabras]
stemmed_words
```

#### _Stemming_ en español (entre otros idiomas)

El paquete `nltk` contiene [SnowballStemmer](https://www.nltk.org/api/nltk.stem.SnowballStemmer.html), que permite convservar las raíces de otros idiomas.

```python
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer("spanish")

texto = 'son las seis de la mañana y me da igual \
voy a salir a la calle, voy a ponerme a gritar \
voy a gritar que te quiero, que te quiero de verdad \
con esa sonrisa puesta, de verdad que no me cuesta \
pensar en ti cuando me acuesto \
pero, Aitana, no imagines el resto \
que si no, no queda bonito esto.'

palabras = word_tokenize(texto)

stemmed_words = [stemmer.stem(palabra) for palabra in palabras]
stemmed_words # ¡SAL NO ES LO MISMO QUE SALIR!
```

#### n-Gramas

```python
from nltk import ngrams
```

```python
frase = 'eso eso eso es todo amigos'

n = 2 # un bigrama
bigramas = ngrams(frase.split(), n)
for gramas in bigramas:
    print (gramas)
```

```python
import networkx as nx

lista_bigramas = list(ngrams(frase.split(), n))

G = nx.DiGraph()

for gram in lista_bigramas:
    G.add_edge(gram[0], gram[1])

plt.figure(figsize=(10, 6))
nx.draw(G, with_labels=True, node_size=3000, font_size=14)
plt.title("Grafo de Bigramas")
plt.show()
```
